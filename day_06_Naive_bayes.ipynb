What is Naïve Bayes?

Naïve Bayes is a method used for making predictions, like guessing if something will happen or not, based on past data.
A simple, linear model that assumes that features are independent of each other.

Why is it called "naïve"? Because it assumes that all features (data points) are independent of each other. So, one feature doesn’t affect another. This is a bit of a “naïve” idea since, in real life, features might be related, but it makes things simpler and faster to calculate.

How Does Naïve Bayes Work?
Uses Bayes' theorem to invert conditional probabilities and calculate the probability of an outcome given some evidence.
Naïve Bayes works on the idea of probability. Probability means the chance of something happening. With Naïve Bayes, we calculate the chance (probability) of something happening based on previous examples.
Naive Bayes is a probabilistic classifier that uses Bayes' theorem to classify events.
When it's useful
Works well for large datasets, binary and multi-class classification, and tasks like spam filtering, text classification, and recommendation systems.

How it's used
For example, in email filtering, Bayesian algorithms have been used since 1996, and many modern mail clients use Bayesian spam filtering.

Let’s use an example:
Imagine you have some emails, and you want to predict if a new email is spam or not.

Step 1: Collect Data
First, look at the data. In this case, the data is words in emails. Words like “buy now” or “limited offer” are more common in spam emails.

Step 2: Calculate Probabilities
We find out how likely it is for certain words to show up in spam emails. For example, if the word “offer” shows up a lot in spam emails, we calculate that probability.

Step 3: Multiply Probabilities
When you get a new email, look at each word’s chance of being in a spam email, and multiply these probabilities together. If the result is high, it’s probably spam; if it’s low, it’s probably not spam.

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer

# Sample data
emails = ["buy blushes at new prizes", "limited time offer, buy now", "meeting tomorrow", "hello we need to meet soon"]
labels = [1, 1, 0, 0]  # 1 = Spam, 0 = Not Spam

# Step 1: Vectorize the emails (Convert words into numbers)
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(emails)

# Step 2: Train Naïve Bayes Classifier
model = MultinomialNB()
model.fit(features, labels)

# Step 3: Predict if a new email is spam or not
test_email = ["get cheap deals now"]
test_features = vectorizer.transform(test_email)
prediction = model.predict(test_features)

print("Is it spam?" if prediction[0] == 1 else "It is not spam")
```


Great for Text: It’s often used for text classification, like detecting spam emails, analyzing sentiments, etc.

**Advantages of Naive Bayes Classifier**

Easy to implement and computationally efficient.

*   Effective in cases with a large number of features.
*   Performs well even with limited training data.
*   It performs well in the presence of categorical features.
*   For numerical features data is assumed to come from normal distributions.

***Disadvantages of Naive Bayes Classifier ***
* Assumes that features are independent, which may not always hold in
real-world data.

*   Can be influenced by irrelevant attributes.
*   May assign zero probability to unseen events, leading to poor generalization.
